```python
from nltk import word_tokenize,sent_tokenize
from nltk.corpus import stopwords

stop_words=set(stopwords.words("english"))
print(stop_words)

```

```python
setence= "Stopwords are a pretty cool way of getting rid of the words that we do not want"
stop_words=set(stopwords.words("english"))
print(stop_words)
```

```python
tokenize_words=word_tokenize(setence)
filtered_sentence=[]
for words in tokenize_words:
  if words not in stop_words:
    filtered_sentence.append(words)

print(filtered_sentence)    
```

```python
with open("pos.txt","r") as corpora:
  text=corpora.read()
```

```python
tokenized_sentence=sent_tokenize(text)
for sentence in tokenized_sentence:
    tokenized_words = word_tokenize(sentence)

    filtered_sentence = []

    for word in tokenized_words:
        
        if word not in stop_words:
            filtered_sentence.append(word)

    print(filtered_sentence)
```
